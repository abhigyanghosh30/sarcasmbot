Getting most common part of speach for the word

```python
def get_part_of_speech(word):
    all_synsets = wn.synsets(word)
    most_common_pos,most_common_name = all_synsets[0].lexname().split(".")
    return (most_common_pos)
```

Generate an indefinite article for the given phrase: a, an, or empty string

```python
def indefinite_article(w):

    if (len(w) == 0):
        return ""
    if w.lower().startswith("a ") or w.lower().startswith("an ") or w.lower().startswith("the "):
        return ""
    return "an " if w.lower()[0] in list('aeiou') else "a "
```

Removing the article in the beginning of the given phrase

```python
def removeArticle(s):

    if s.startswith("a "):
        return s[2:]
    elif s.startswith("an "):
        return s[3:]
    elif s.startswith("the "):  
        return s[4:]
    return s 
```

Camel Case the given string

```python
def camel(s):
    """Camel case the given string"""
    return s[0].upper() + s[1:]
```

Type 1 joke formatting and Generation

```python
def Type1_Joke(d1, d2, w1, w2):
    
    d1 = indefinite_article(d1) + d1
    d2 = removeArticle(d2) + "? "
    return "What do you call " + d1 + " that is  " + d2 + camel(indefinite_article(w1)) + w1 + " " + w2 + "."
```
